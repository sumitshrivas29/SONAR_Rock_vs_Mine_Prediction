# SONAR Rock vs Mine prediction project

### Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### loading the data and Data processing

```python
df= pd.read_csv("D:\other ml projects\Copy of sonar data.csv",header=None)
```

```python
# Splitting the data
x=df.drop(60,axis=1)
y=df[60]
```

```python
x.shape
```

```python
y.shape
```

#### Training and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.1, random_state=1)
```

#### Model Training

```python
model= LogisticRegression()
```

```python
#Training the model
model.fit(x_train,y_train)
```

```python
#accuracy on training data

train_pred= model.predict(x_train)
accuracy1=accuracy_score(train_pred,y_train)
print(accuracy1)
```

```python
#Accuracy on test data

test_pred=model.predict(x_test)
accuracy2=accuracy_score(test_pred,y_test)
print(accuracy2)
```
